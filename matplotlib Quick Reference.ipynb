# Matplotlib

matplotlib.pyplot has a bunch of commands that works exactly like MATLAB.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4, 6, 4])
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

```python
a = np.arange(0,6,1)
plt.plot(a,a**2)
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # grid on

plt.show()
```

```python
a = np.arange(0,6,0.1)
plt.plot(a,a**2)
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # grid on

plt.show() # more smooth
```

```python
a = np.arange(0,6,1)
plt.plot(a,a**2,'ro')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # grid on

plt.show()
```

```python
a = np.arange(0,6,1)
plt.plot(a,a**2,'ro')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()

plt.show()
```

```python
a = np.arange(0,6,0.25)
plt.plot(a, a**2, 'r^', label = 'Power of 2')
plt.plot(a, a**2.25, 'gs', label = 'Power of 2.25')
plt.plot(a, a**2.5, 'b--', label = 'Power of 2.5')

plt.legend()  # label on the plot
```

```python
x1 = [1, 2, 3, 4]
y1 = [i**2 for i in x1]
x2 = [1, 2, 3, 4]
y2 = [i*2 for i in x2]

plt.plot(x1,y1, 'r', linewidth = 3)

plt.plot(x2,y2, 'g', linewidth = 3)

plt.xticks(x1)
plt.yticks(np.arange(0,17,4))

plt.grid()
```

```python
t1 = np.linspace(-2*np.pi, 2*np.pi)               


plt.figure(1)

# The subplot() command specifies numrows, numcols, 
# subplot(211) 2 implies 2 rows, 1 implies 1 column, 1 implies plot number 1

plt.subplot(211)
plt.grid()
plt.title('Plot 1')
plt.plot(t1, np.sin(t1) , 'b-')

# subplot(211) 2 implies 2 rows, 1 implies 1 column, 1 implies plot number 2
plt.subplot(212)
plt.title('Plot 2')
plt.plot(t1, np.cos(t1), 'r--')
plt.show()
```
